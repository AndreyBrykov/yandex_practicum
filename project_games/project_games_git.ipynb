Проект выполнен на основе данных интернет-магазина «Стримчик», который продаёт по всему миру компьютерные игры. Для анализа были доступны исторические данные о продажах игр, оценке пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).

Целью проекта является выявление определяющих успешность игры закономерностей, что позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

```python
# выведем уникальные значения
print(data['user_score'].unique())
# посчитаем сколько раз встречается отличное от других значение - 'tbd'
data[data['user_score']=='tbd']['user_score'].count()
```

Очевидно, что `tbd` - распространенная английская абревиатура 'To Be Determine' (подлежит определению). Раз так, то в рамках нашего исследования можно не различать `tbd` и `NaN` приведя тип данных к вещественному, что позволит в полной мере использовать его преимущества. Для себя отметим более 6 тыс. записей об играх не имеют оценки пользователей. Заменим `tbd` на `NaN` и изменим тип данных на `float32`.

```python
# меняем tbd на NaN
data.loc[data.loc[:,'user_score']=='tbd','user_score'] = np.nan
# меняем тип данных
data['user_score'] = data['user_score'].astype('float32')
```

#### Работа с пропусками

Обзор таблицы показал, что подавляющее большинство столбцов с наименованием, жанром игр и информацией о продажах заполнены данными, тогда как столбцы с оценками пользователей и экспертов - `critic_score`, `user_score`, `rating` имеют много пропусков.
Обработуем пропуски.

**Наименование игры `name`** - отсутствующее название игры не позволит нам сделать верные суждения, применимые в целях настоящего исследования. Коммерческое наименование игры должно быть, хоть оно и может быть "Без названия", но пропуск вероятно вызван технической ошибкой. Посмотрим на эти строки и убедимся, что это не топовые по продажам игры и их удаление не окажет существенного влияния на результаты.

```python
# игры без названий
data[data['name'].isna()].reset_index(drop=True)
```

```python
# удаляем строки без названий игр
data = data.dropna(subset = ['name'])
```

**Год выпуска игры `year_of_release`** - эта информация обычно известна и ее отсуствие кроме как технической ошибкой сбора или выгрузки информации объяснить сложно. Если игра до сих пор популярна, то даже если ее выпустили давно на нее можно делать ставку в продажах. В рамках исследования наличие года выпуска игры имеет значение, но часть исследований мы сможем сделать и без этой информации. Учитывая что `NaN` относится к типу `float` оставим пропуски без изменений для удоства дальнейшего анализа.

**Оценка критиков `critic_score`**  - вероятно, пропуски здесь это игры, котрые еще не оценены критиками. Ввиду большого числа пропусков удалять строки не следует. Заполнять расчетным значением медианы будет не корректным ввиду возможного влияния на корреляцию с другими данными. Вместо этого заполним пустые значения значением "-1", таким образом сохранив тип даных в столбце и обозначив пустые значения для анализа.

Аналогично поступим со столбцом **Оценка пользователей** - `user_score` 

```python
data['critic_score'] = data['critic_score'].fillna(-1)
data['user_score'] = data['user_score'].fillna(-1)
```

Пустые значения в столбце **Рейтинг ESRB** (англ. Entertainment Software Rating Board) - `rating` заполним значением "unknown". Возможно, это игры, рейтинг которым еще не присвоен. Кроме того, существуют альтернативные классификации, данных о которых у нас нет.

```python
data['rating'] = data['rating'].fillna('unknown')
# посмотрим что теперь с заполненностью таблицы
pd.DataFrame(round(data.isna().mean().sort_values()*100,5)).style.background_gradient('coolwarm')
```

```python
# проверим так ли это - есть ли одинаковые названия и платформы, но релизы в разные годы?
data[data.duplicated(['name','platform'], keep=False)].sort_values(['name','platform']).reset_index(drop=True)
```

### Исследовательский анализ данных

#### Время жизни платформ, актуальный период и лидеры продаж

Сравним число игр, выпущенное в разные годы и определим период, который будет нас интересовать в рамках настоящего исследования.

```python
# сгруппируем данные по годам и вычислим число релизов  в год
data.groupby(by='year_of_release').agg(games_for_year=('name', 'count')).plot(kind='bar', alpha=0.7, 
                                                                               color='g', figsize=(12,5), rot=45)
plt.xlabel('Годы продаж игр')
plt.ylabel('Число релизов')
plt.title('Число релизов в год')
plt.show()
```

Очевидно не всеми годами нам придется воспользоваться. Распространенность и возможности платформ сильно поменялись по прошествии десятилетия. В этой связи данные до 2001 года в анализе нам не будут полезны. Текущий тренд в продажах, предпочтениях пользователей и современные возможности платформ говорят нам о том, что стоит ориентироваться на последние 3-5 лет в анализе. В тоже время закономерности, которые определяют успешность игры могут быть более устойчивыми и сохраняться на протяжениии нескольких лет. Позднее, изучив время жизни различным игровых платформ, мы подробнее убедимся на какой период опираться в анализе при рассмотрении тех или иных показателей. Учитывая тенденции начала 2000-х и заметное снижение числа релизов после 2010, подробней рассмотрим именно этот период, который в большей мере отражает современные реалии.

```python
# сгруппируем платформы и вычислим их годы использования - период релизов игр на них
data_platform_lifetime = data.pivot_table(index='platform', values= 'year_of_release', aggfunc=['min','max'])
data_platform_lifetime = data_platform_lifetime.reset_index()
data_platform_lifetime.columns = ['platform', 'start', 'finish']
data_platform_lifetime['lifetime'] = data_platform_lifetime['finish'] -  data_platform_lifetime['start']
# первые 5 строк
data_platform_lifetime.head(5)
```

#### Влияние отзывов на продажи игр по платформам

Посмотрим как влияют на продажи отзывы пользователей. Построим диаграммы рассеяния между отзывами пользователей и продажами. Сделаем это сразу для ТОП-5 самых продаваемых платформ в актуальный период. Вычислим для них коэффициент корреляции.

```python
# пробегаем циклом по самым продаваемым платформам в актуальный период
for val in max_sales_platform_2013.index:
    # делаем срез данных по конкретной платформе с наличием оценок пользователей
    # и строим диаграмму рассеяния
    data_actual.query('platform==@val & user_score>=0').plot(kind='scatter', 
                                                                  x='user_score', 
                                                                  y='total_sales', 
                                                                  figsize=(5,5), color='g', alpha=0.4,
                                                                  label=f'платформа: {val}')
    # задаем координаты линии тренда и выводим ее на график
    x1=data_actual.query('platform==@val & user_score>=0')['user_score']
    y1=data_actual.query('platform==@val & user_score>=0')['total_sales']
    z = np.polyfit (x1, y1, 1)
    p = np.poly1d (z)
    plt.plot(x1, p(x1), color='r')   
    plt.xlabel('Оценка пользователей')
    plt.ylabel('Продажи в мире, млн копий')
    plt.title('Диаграмма рассеяния (оценка пользователей / продажи)')
    plt.show()
    # посчитаем коэффициент корреляции между продажами и отзывами пользователей
    (print(f'Коэффициент корреляции продаж и отзывов пользователей для платформы {val}:',
           round(data_actual.query('platform==@val & user_score>=0')['total_sales']
           .corr(data_actual.query('user_score>=0')['user_score']),2))
    )
```

Посмотрим как влияют на продажи отзывы критиков. Построим диаграммы рассеяния между отзывами критиков и продажами сразу для ТОП-5 платформ по продажам в актуальный период.  Вычислим для них коэффициент корреляции.

```python
# пробегаем циклом по самым продаваемым платформам в актуальный период
for val in max_sales_platform_2013.index:
    # делаем срез данных по конкретной платформе с наличием оценок критиков
    # и строим диаграмму рассеяния
    data_actual.query('platform==@val & critic_score>=0').plot(kind='scatter', 
                                                                  x='critic_score', 
                                                                  y='total_sales', 
                                                                  figsize=(5,5), color='b', alpha=0.4,
                                                                  label=f'платформа: {val}')
    # задаем координаты линии тренда и выводим ее на график    
    x1=data_actual.query('platform==@val & critic_score>=0')['critic_score']
    y1=data_actual.query('platform==@val & critic_score>=0')['total_sales']
    z = np.polyfit (x1, y1, 1 )
    p = np.poly1d (z)
    plt.plot (x1, p(x1), color='r') 
    plt.xlabel('Оценка критиков')
    plt.ylabel('Продажи в мире, млн копий')
    plt.title('Диаграмма рассеяния (оценка критиков / продажи)')
    plt.show()
    plt.show()
    # посчитаем коэффициент корреляции между продажами и отзывами критиков
    (print(f'Коэффициент корреляции продаж и отзывов критиков для платформы {val}:',
           round(data_actual.query('platform==@val & critic_score>=0')['total_sales']
           .corr(data_actual.query('critic_score>=0')['critic_score']),2))
    )
```

**Выводы:**

Результаты анализа показали отсутствие связи между продажами и отзывами пользователей для всех платфом, за исключением очень слабой связи для `3DS`, которая не позволяет говорить о зависимости.

В тоже время, для этих же платформ корреляция продаж и отзывов критиков выражена в большей степени и имеет положительную зависимость для всех платформ. Но эту связь также можно охарактеризовать как слабовыраженную.

### Проверка гипотез

Проверим несколько гипотез относительно рейтингов игр.

#### Средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые?

Сформулируем нулевую гипотезу **H0** следующим образом: средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые.
Альтернативную гипотезу **H1** сформулируем следующим образом: средние пользовательские рейтинги платформ `Xbox One` и `PC` различны.

Для проверки гипотезы сделаем срез в данных актуального периода по сравниваемым платформам. Проверим объем этих совокупностей:

```python
# для корректной проверки гипотез убедимся в достаточности данных
print('Число игр на платформе XOne с оценками пользователей:', 
      data_actual.query('platform == "XOne" & user_score>=0')['user_score'].count())
print('Число игр на платформе PC с оценками пользователей:', 
      data_actual.query('platform == "PC" & user_score>=0')['user_score'].count())
```

Отметим, что данные в полученных совокупностях будут независимыми.

Кроме того, согласно центральной предельной теоремы и учитывая объем выборок, среднее каждой из этих выборок будет нормально распределено вокруг истинного среднего генеральной совокупности.

Поскольку мы не обладаем полной иформацией о дисперсиях сравниваемых выборок, то параметр `equal_var` зададим значением `False`.

```python
# зададим уровень статистической значимости
alpha = .05 
# получим результат сравнения средних
results = st.ttest_ind(data_actual.query('platform == "XOne" & user_score>=0')['user_score'],
                      data_actual.query('platform == "PC" & user_score>=0')['user_score'],
                      equal_var = False)
# выведем вероятность полученного или большего отклонения средних сравниваемых выборок между собой
print('p-значение:', round(results.pvalue,5))

# задаем критерий проверки гипотез
# нулевая гипотеза будет отвергнута, если вероятность получить такое или большее отклонение в средних 
# не превысит уровень статистической значимости
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')

# посмострим на значения средних пользовательских рейтингов сравниваемых платформ
print('Средняя пользовательская оценка игр на платформе XOne:', 
      round(data_actual.query('platform == "XOne" & user_score>=0')['user_score'].mean(),2))
print('Средняя пользовательская оценка игр на платформе PC:', 
      round(data_actual.query('platform == "PC" & user_score>=0')['user_score'].mean(),2))
```

**Вывод:** в результате статистического сравнения мы не смогли отвергнуть гипотезу о равенстве средних пользовательских рейтингов платформ `Xbox One` и `PC`.

#### Средние пользовательские рейтинги жанров `Action` и `Sports` разные?

Сформулируем нулевую гипотезу **H0** следующим образом: средние пользовательские рейтинги жанров `Action` и `Sports` одинаковые.
Альтернативную гипотезу **H1** сформулируем следующим образом: средние пользовательские рейтинги жанров `Action` и `Sports` различны.

```python
# для корректной проверки гипотез убедимся в достаточности данных
print('Число игр жанра Action с оценками пользователей:', 
      data_actual.query('genre == "Action" & user_score>=0')['user_score'].count())
print('Число игр жанра Sports с оценками пользователей:', 
      data_actual.query('genre == "Sports" & user_score>=0')['user_score'].count())
```
